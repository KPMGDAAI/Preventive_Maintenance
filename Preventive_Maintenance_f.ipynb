# Purpose of windoing and overlapping in Vibration Analysis
Vibration analysis often involves examining time-series data collected from sensors monitoring vibrations in machinery or structures. Analyzing the entire dataset at once might not reveal specific patterns or anomalies. By segmenting the data into smaller windows, you can analyze these segments individually to detect changes over time, identify trends, or spot anomalies that might indicate issues like wear, imbalance, or misalignment in machinery.

```python
def create_windows(data, window_size, overlap):
    step_size = int(window_size * (1 - overlap))  # Calculate step size based on overlap
    windows = []

    for start in range(0, len(data) - window_size + 1, step_size):
        window = data[start:start + window_size]
        windows.append(window)

    return np.array(windows)
```


# Explanation of extract_features Function

The extract_features function calculates key statistical features from windows of vibration data, encapsulating the main characteristics of the vibration signal in each segment. These features are crucial for further analysis, enabling more effective monitoring and diagnosis of potential issues in machinery based on vibration data.

```python
def extract_features(windows):
    features = []

    for window in windows:
        # Calculate statistical features for each window
        mean = np.mean(window)
        std_dev = np.std(window)
        max_value = np.max(window)
        min_value = np.min(window)
        skewness = skew(window)
        kurt = kurtosis(window)

        # Append features to the list
        features.append([mean, std_dev, max_value, min_value, skewness, kurt])

    # Convert to a DataFrame for easier manipulation
    columns = ['Mean', 'Std_Dev', 'Max', 'Min', 'Skewness', 'Kurtosis']
    return pd.DataFrame(features, columns=columns)
```

# Combine Feature DataFrames into a Single DataFrame

combined_features: A DataFrame that combines all feature DataFrames into one.
pd.concat(features_dfs.values(), axis=1): Concatenates the feature DataFrames along columns.
combined_features.reset_index(drop=True, inplace=True): Resets the index of the combined DataFrame and drops the old index, ensuring a clean sequential index.

```python
# Parameters
window_size = 5
overlap = 0.5

# Create windows for each feature
windowed_data = {}
for column in df.columns:
    if column != 'Date_time':  # Exclude 'Date_time' column
        data = df[column].values
        windows = create_windows(data, window_size, overlap)
        windowed_data[column] = windows

# Extract features for each windowed column
features_dfs = {}
for column, windows in windowed_data.items():
    features_dfs[column] = extract_features(windows)

# Combine feature DataFrames into a single DataFrame
combined_features = pd.concat(features_dfs.values(), axis=1)
combined_features.reset_index(drop=True, inplace=True)
```

# Anomaly Detection and Feature Processing

![alt tab](https://lh3.google.com/u/0/d/1SGn0eYh0khqZRBqwUm7l6RI4Z8YLKiS5=s320)


This code block processes a list of DataFrames representing different feature sets from vibration data. It performs anomaly detection by training an Isolation Forest model and imputes missing values.


**Define DataFrames and Names:** The dataframes list contains the DataFrames X_Axis_DZ, Z_Axis_DZ, X_Axis_CZ2, and Z_Axis_CZ2. The df_names list holds corresponding names for these DataFrames.


**Feature Selection and Imputation:** For each DataFrame in the list:

**Select Features:** Relevant statistical features (Mean, Std_Dev, Max, Min, Skewness, Kurtosis) are extracted from the DataFrame.
**Impute Missing Values:** A SimpleImputer with a mean strategy is used to replace any missing values in the selected features with the mean of each column.
**Train Isolation Forest:** An Isolation Forest model is trained on the imputed feature data. This model identifies anomalies by predicting the anomaly scores for each feature set. The contamination parameter is set to 0.05, indicating the expected proportion of outliers.

**Add Anomaly Scores:** The predicted anomaly scores are added to the DataFrame as a new column, Anomaly_Score. This column indicates the likelihood of anomalies in the data based on the Isolation Forest model.

```python


# List of DataFrames
dataframes = [X_Axis_DZ, Z_Axis_DZ, X_Axis_CZ2, Z_Axis_CZ2]
df_names = ['X_Axis_DZ', 'Z_Axis_DZ', 'X_Axis_CZ2', 'Z_Axis_CZ2']

# Loop through each DataFrame
for i, df in enumerate(dataframes):
    # Select the features
    features = df[['Mean', 'Std_Dev', 'Max', 'Min', 'Skewness', 'Kurtosis']]

    # Impute missing values with the mean of each column
    imputer = SimpleImputer(strategy='mean')
    features_imputed = imputer.fit_transform(features)

    # Train Isolation Forest
    iso_forest = IsolationForest(contamination=0.05, random_state=42)
    anomaly_scores = iso_forest.fit_predict(features_imputed)

    # Add anomaly scores to the DataFrame
    df['Anomaly_Score'] = anomaly_scores


```

# Control Limits Calculation
This script is designed to calculate control limits for various statistical features of data collected from different axes, applying the 3-sigma rule to identify the acceptable range of variability. Here's a detailed breakdown of the process:

**DataFrames Initialization:** The script begins with a dictionary named dataframes, which contains several DataFrames, each corresponding to a different axis of measurement. These DataFrames are labeled as X_Axis_DZ, Z_Axis_DZ, X_Axis_CZ2, and Z_Axis_CZ2.

**Control Limits Initialization:** An empty dictionary called control_limits_all_axes is created to store the calculated control limits for each axis.

3-Sigma Rule Application: The variable k is set to 3, representing the 3-sigma rule. This rule is used to calculate the control limits by determining the mean and standard deviation of each statistical feature and then setting the control limits as ±3 standard deviations from the mean.

**Processing Each DataFrame:** The script loops through each DataFrame in the dataframes dictionary. For each DataFrame:

**Mean and Standard Deviation Calculation:** The script calculates the mean and standard deviation for several statistical features: 'Mean', 'Std_Dev', 'Max', 'Min', 'Skewness', and 'Kurtosis'. This is done using the agg method to get aggregated statistics.
Control Limits Calculation: For each feature, it computes the Lower Control Limit (LCL) and Upper Control Limit (UCL) using the formula:

Storing Control Limits: The calculated control limits for each feature are stored in a dictionary and then added to the main control_limits_all_axes dictionary.
Combining Results: After processing all DataFrames, the individual control limits DataFrames are concatenated into a single DataFrame named all_control_limits_df. This consolidated DataFrame provides a comprehensive view of the control limits across all axes and features.

**Output:** Finally, the script prints the combined DataFrame of control limits, showing the calculated LCL and UCL for each feature across different measurement axes.

This process helps in monitoring and ensuring the consistency and quality of the data by identifying potential deviations from expected variability, thereby supporting effective data control and decision-making.

```python


# Assuming you have the following DataFrames
dataframes = {
    'X_Axis_DZ': X_Axis_DZ,
    'Z_Axis_DZ': Z_Axis_DZ,
    'X_Axis_CZ2': X_Axis_CZ2,
    'Z_Axis_CZ2': Z_Axis_CZ2
}

# Initialize an empty dictionary to hold control limits for all axes
control_limits_all_axes = {}

# 3-sigma rule
k = 3

# Loop through each DataFrame and calculate control limits
for axis_name, df in dataframes.items():
    # Calculate mean and standard deviation for each feature
    mean_std = df.agg({
        'Mean': ['mean', 'std'],
        'Std_Dev': ['mean', 'std'],
        'Max': ['mean', 'std'],
        'Min': ['mean', 'std'],
        'Skewness': ['mean', 'std'],
        'Kurtosis': ['mean', 'std']
    })

    # Calculate control limits for each feature
    control_limits = {}
    for feature in ['Mean', 'Std_Dev', 'Max', 'Min', 'Skewness', 'Kurtosis']:
        mean = mean_std.loc['mean', feature]
        std_dev = mean_std.loc['std', feature]

        control_limits[feature] = {
            'LCL': mean - k * std_dev,
            'UCL': mean + k * std_dev
        }

    # Store the control limits in the main dictionary
    control_limits_all_axes[axis_name] = pd.DataFrame(control_limits)

# Concatenate all control limits DataFrames into a single DataFrame
all_control_limits_df = pd.concat(control_limits_all_axes, axis=0)

print(all_control_limits_df)
```

# Creating Labels for the anomolies to set as Target Variable to feed into the Regression Model

```python
anomaly_labels = []

# Iterate over each row in the DataFrame
for index, row in combined_anomaly_df.iterrows():
    # Count the number of -1 values in the current row
    num_anomalies = np.sum(row == -1)

    # Determine the label based on the number of anomalies
    if num_anomalies == 4:
        label = 4
    elif num_anomalies == 3:
        label = 3
    elif num_anomalies == 2:
        label = 2
    elif num_anomalies == 1:
        label = 1
    else:
        label = 0

    # Append the label to the list
    anomaly_labels.append(label)

# Add the labels as a new column to the DataFrame
combined_anomaly_df['Anomaly_Label'] = anomaly_labels
```

```python
# Extract the first 6 columns from each DataFrame
X_Axis_DZ_features = X_Axis_DZ.iloc[:, :-1]
Z_Axis_DZ_features = Z_Axis_DZ.iloc[:, :-1]
X_Axis_CZ2_features = X_Axis_CZ2.iloc[:, :-1]
Z_Axis_CZ2_features = Z_Axis_CZ2.iloc[:, :-1]

# Combine features into a single DataFrame
combined_features = pd.concat([
    X_Axis_DZ_features,
    Z_Axis_DZ_features,
    X_Axis_CZ2_features,
    Z_Axis_CZ2_features
], axis=1)



# Combine features with labels
final_df = pd.concat([combined_features, combined_anomaly_df['Anomaly_Label']], axis=1)

print(final_df)
```

# Regression Modeling
This code snippet performs several key steps in building and evaluating a logistic regression model for anomaly detection using a dataset:

Feature and Label Selection: The DataFrame df is used to extract features and labels. The features used for model training are Mean, Std_Dev, Max, Min, Skewness, and Kurtosis, stored in X. The target label, which indicates whether an anomaly is present, is stored in y as Anomaly_Label.

**Handling Missing Values: **Missing values in the feature set X are addressed using mean imputation. The SimpleImputer is used to replace any missing values with the mean of each feature, ensuring that the dataset is complete and ready for model training.

**Label Encoding:** The categorical anomaly labels in y are encoded into numerical format using LabelEncoder. This converts the labels into a format suitable for machine learning algorithms, where each label is assigned a unique integer.

**Data Splitting: **The dataset is split into training and testing sets using train_test_split. The training set, comprising 80% of the data, is used to train the model, while the testing set, comprising 20% of the data, is used to evaluate the model's performance. The random_state is set to ensure reproducibility of the results.

**Model Initialization and Training:** A logistic regression model is initialized with the option multi_class='ovr' (one-vs-rest strategy for multi-class classification) and max_iter=1000 to specify the maximum number of iterations for optimization. The model is then trained using the training data.

**Making Predictions:** The trained model is used to make predictions on the test set, generating predicted labels for comparison with the true labels.

**Model Evaluation:** The accuracy of the model is computed by comparing the predicted labels to the actual labels in the test set. Additionally, a detailed classification report is generated, which includes metrics such as precision, recall, and F1-score for each class. The target_names are set to the original class labels (converted to strings), and zero_division=0 ensures that the report handles cases where there might be zero predicted instances for a class.

Overall, the code processes the data, trains a logistic regression model, and evaluates its performance in predicting anomalies, providing insights into how well the model can distinguish between normal and anomalous instances.

```python


# Assuming df is your DataFrame
df = final_df

# Features and labels
X = df[['Mean', 'Std_Dev', 'Max', 'Min', 'Skewness', 'Kurtosis']]
y = df['Anomaly_Label']

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_encoded, test_size=0.2, random_state=42)

# Initialize and train the logistic regression model
model = LogisticRegression(multi_class='ovr', max_iter=1000)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))

# Convert class labels to string for classification_report
target_names = [str(cls) for cls in label_encoder.classes_]

print("Classification Report:\n", classification_report(y_test, y_pred, target_names=target_names, zero_division=0))
```
